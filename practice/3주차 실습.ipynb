# Assignment 2: 사용자별로 처음과 마지막 채널 알아내기

ROW_NUMBER

```python
%%sql
WITH tmp_tb AS (
    SELECT userid, ts, channel,
    ROW_NUMBER() OVER (partition by userid order by ts) seq,
    ROW_NUMBER() OVER (partition by userid order by ts desc) seq_rev
    FROM raw_data.user_session_channel usc
    JOIN raw_data.session_timestamp st ON usc.sessionid = st.sessionid
)
SELECT first_tb.userid, firstchannel, lastchannel
FROM
  (SELECT userid, channel firstchannel
  FROM tmp_tb
  WHERE tmp_tb.seq = 1) first_tb
  JOIN
  (SELECT userid, channel lastchannel
  FROM tmp_tb
  WHERE tmp_tb.seq_rev = 1) last_tb
  ON first_tb.userid = last_tb.userid
ORDER BY 1
```

FIRST_VALUE/LAST_VALUE

```python
%%sql

SELECT DISTINCT userid,
FIRST_VALUE(channel) OVER (partition by userid order by ts rows between unbounded preceding and unbounded following) firstchannel,
LAST_VALUE(channel) OVER (partition by userid order by ts rows between unbounded preceding and unbounded following) lastchannel
FROM raw_data.user_session_channel usc
JOIN raw_data.session_timestamp st ON usc.sessionid = st.sessionid
ORDER BY 1
```

# Assignment 3: Gross Revenue가 가장 큰 UserID 10개 찾기

```python
%%sql

SELECT userid, SUM(amount) "Gross Revenue"
FROM raw_data.user_session_channel usc
JOIN raw_data.session_transaction st ON usc.sessionid = st.sessionid
GROUP BY 1
ORDER BY 2 desc
LIMIT 10

```

# Assignment 4: 채널별 월 매출액 테이블 만들기



```python
%%sql

DROP TABLE IF EXISTS dlstlr28.chnnal_month_sales;

CREATE TABLE dlstlr28.chnnal_month_sales AS
SELECT TO_CHAR(ts, 'YYYY-MM') AS month
    , channel
    , COUNT(DISTINCT userid) uniqueUsers
    , COUNT(sta.amount) paidUsers
    , TRUNC(paidUsers / uniqueUsers::NUMERIC, 2) conversionRate
    , SUM(amount) grossRevenue
    , SUM(CASE WHEN refunded = True THEN 0 ELSE amount END) netRevenue
FROM raw_data.user_session_channel usc
LEFT JOIN raw_data.session_timestamp stm ON usc.sessionid = stm.sessionid
LEFT JOIN raw_data.session_transaction sta ON stm.sessionid = sta.sessionid
GROUP BY 1, 2
ORDER BY 1 desc, 3 desc;

SELECT *
FROM chnnal_month_sales
```
