# Assignments 3: Summary table을 빌드해보기

```python
from airflow import DAG
from airflow.operators import PythonOperator
from airflow.models import Variable
from airflow.hooks.postgres_hook import PostgresHook
from datetime import datetime
from datetime import timedelta
import requests
import logging
import psycopg2

def get_Redshift_connection():
    # Airflow web UI 에 등록된 Connection id
    hook = PostgresHook(postgres_conn_id='redshift_dev_db')
    return hook.get_conn().cursor()


def execSQL(**context):
    schema = context['params']['schema'] 
    table = context['params']['table']
    select_sql = context['params']['sql']

    cur = get_Redshift_connection()

    try:
        sql = "BEGIN;"
        sql += f"DROP TABLE IF EXISTS {schema}.{table};"
        sql += f"CREATE TABLE {schema}.{table} AS {select_sql};"
        sql += "END;"
        logging.info(sql)
        cur.execute(sql)
    except Exception as e:
        cur.execute("ROLLBACK;")
        logging.error("Failed to create summary table.")

dag = DAG(
        dag_id = "ctas_summary_table",
        start_date = datetime(2021, 5, 20),
        schedule_interval = '@once',
        catchup = False)

execsql = PythonOperator(
        task_id = 'execsql',
        python_callable = execSQL,
        params = {
            'schema' : 'analytics',
            'table': 'FirstLastChannel',
            'sql' : '''
                    SELECT DISTINCT A.userid,
                           FIRST_VALUE(A.channel) over(partition by A.userid order by B.ts rows between unbounded preceding and unbounded following) AS first_channel,
                           LAST_VALUE(A.channel) over(partition by A.userid order by B.ts rows between unbounded preceding and unbounded following) AS last_chaneel
                    FROM dlstlr28.user_session_channel AS A
                    LEFT JOIN dlstlr28.session_timestamp AS B
                    ON A.sessionid = B.sessionid;
                    '''
        },
        provide_context = True,  # 파라미터 전달을 위한 설정
        dag = dag
        )

execsql
```

```python
%%sql
--월별 나이대의 총 구매량, 클릭 후 구매하지 않은 품목 수

SELECT datestamp, engagement_level,
       SUM(CASE WHEN purchased = 1 THEN price * paidamount ELSE 0 END) amount,
       COUNT(CASE WHEN (clicked = 1 and purchased = 0) THEN 1 END) AS clicked_only
FROM dlstlr28.customer_features AS cf
JOIN dlstlr28.customer_interactions AS ci ON cf.customer_id = ci.customer_id
JOIN dlstlr28.item_features AS itf ON ci.item_id = itf.item_id
GROUP BY 1, 2
ORDER BY 1, 2
```
